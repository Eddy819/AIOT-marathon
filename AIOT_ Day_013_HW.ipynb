# 作業二
* 實際練習GPIOZero控制LED，確定單獨的GPIO控制LED亮跟暗交替閃爍完成，並且PWM控制LED明亮的完成後，嘗試依序改變led.value的值，分別設定0.1, 0.3, 0.5, 0.7觀察差異。

```python
from gpiozero import PWMLED
from time import sleep

# sleepInterval 每個指令之間的等待時間
sleepInterval = 1
# 定義使用GPIO 17作為GPIO的接腳
led = PWMLED(17)
while True:
    led.value = 0.1
    sleep(sleepInterval)
    led.value = 0.3
    sleep(sleepInterval)
    led.value = 0.5
    sleep(sleepInterval)
    led.value = 0.7
    sleep(sleepInterval)
```

# 作業三
* 實際練習GPIOZero透過Button控制LED，在按鈕的過程中，觀察實際按下按鈕的次數，LED點亮的次數，是否一致。練習修改程式，讓按鈕按下是全亮，按鈕放開後是30%的亮度

```python
from gpiozero import LED, Button
from singal import pause

led = LED(17)

def lightOn():
    led.value = 1
    print("Light value is 1")

def lightOff():
    led.value = 0.3
    print("Light value is 0.3")

button = Button(2)
button.when_pressed = lightOn
button.when_released = lightOff

pause()
```
